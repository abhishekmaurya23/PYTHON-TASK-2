Certainly! Here's a Python function that checks whether a password entered by the user follows the given guidelines:

```python
def check_password(password):
    uppercase_count = 0
    lowercase_count = 0
    number_count = 0
    special_count = 0

    if len(password) != 10:
        return "Invalid Password"

    for char in password:
        if char.isupper():
            uppercase_count += 1
        elif char.islower():
            lowercase_count += 1
        elif char.isdigit():
            number_count += 1
        elif char in "!@#$%^&*":
            special_count += 1

    if uppercase_count >= 2 and lowercase_count >= 2 and number_count >= 1 and special_count >= 3:
        return "Valid Password"
    else:
        return "Invalid Password"


# Test the function
user_password = input("Enter your password: ")
result = check_password(user_password)
print(result)
```

In this program, the `check_password` function takes a password as input. It initializes separate counters for counting uppercase letters (`uppercase_count`), lowercase letters (`lowercase_count`), numbers (`number_count`), and special characters (`special_count`).

The function first checks if the length of the password is exactly 10 characters. If not, it immediately returns "Invalid Password" since the length requirement is not met.

Then, using a `for` loop, it iterates through each character in the password and increments the respective counter based on the character's type (uppercase, lowercase, digit, or special character).

Finally, the function checks if the counts of uppercase letters, lowercase letters, numbers, and special characters meet the required guidelines. If all the conditions are satisfied, it returns "Valid Password". Otherwise, it returns "Invalid Password".

You can test the function by entering a password and checking the result. It will return "Valid Password" if the entered password meets the guidelines, and "Invalid Password" otherwise.
